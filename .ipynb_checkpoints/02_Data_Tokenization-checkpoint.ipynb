### We want to remove bylines from the titles, which indicate news sources. Rather, we can store it in another column to just focus on the title content
Most bylines start with [ - ],[ | ], [; ] 

```python
# Cleaning titles
df['title']=df['title'].apply(lambda x: re.sub(r'(?:[\n]+)','',str(x)))
df['title']=df['title'].apply(lambda x: re.sub(r'(?:[\t]+)','',str(x)))
```

```python
def find_source(x):
    pat_source=r'(?:\s\-[\s]+)([\w\W\d\D\s\S]*)|(?:\s\|[\s]+)([\w\W\d\D\s\S]*)|(?:\;[\s]+)([\w\W\d\D\s\S]*)|(?:\s\–[\s]+)([\w\W\d\D\s\S]*)'
    matches = re.findall(pat_source, x)
    if matches:
        src = ''.join(matches[-1]).strip()
        return src
    else:
        return ""
```

```python
df['article_source']=df['title'].apply(lambda x: find_source(str(x)))
```

```python
df[['title','article_source']].sample(10)
```

```python
df['clean_title']=df['title'].apply(lambda x: re.sub(pat_source,'',str(x)))
df[['title','clean_title','article_source']].sample(10)
```

### Tokenizing Text

```python
def create_tokens(sentence):
    list_tokens=list(gensim.utils.simple_preprocess(str(sentence), deacc=True, min_len=2, max_len=15))  # deacc=True removes punctuations
    list_tokens = [word for word in list_tokens if word not in stop_words]
    return list_tokens
```

```python
df['title_tokens']=df['clean_title'].apply(lambda x: create_tokens(x))
```

```python
df[['title','clean_title','title_tokens']].head(10)
```

```python
import itertools
all_title_tokens=list(itertools.chain(*df['title_tokens']))
all_title_tokens[0:10]
```

```python
fdist_comm_lanc = nltk.FreqDist(all_title_tokens)
fdist_comm_lanc.most_common(20)
```

```python
%%time

# First removing \r - carriage return
df2['cleaned_text']=df2['cleaned_text'].apply(lambda x: re.sub(r'(?:[\r]+)',' ',str(x)))
```

```python
# Start Removing web remnants from the text

words_remove=['AdBlock','Refresh','page','Search','Login','Register','Subscribe','Forgot Username','Forgot Password','About Us','Contact','Privacy Policy',
             'Cookie','Cookies','Cookie Policy','Follow Us','Sign In', 'Inc','T&C', 'Copyright', 'T&CPrivacy',
             'Sign-up','All rights reserved','all rights reserved','All Rights Reserved']
```

```python
%%time
df2['cleaned_text'] = df2['cleaned_text'].str.replace('|'.join(words_remove),'', regex=True)
```

```python
df2['cleaned_text'].head()
```

```python
%%time

df2['cleaned_text'] = df2['cleaned_text'].map(lambda x: re.sub('[^a-zA-Z0-9 @ . , : - _]', '', str(x)))
df2['cleaned_text'] = df2['cleaned_text'].map(lambda x: re.sub('[\s]{2,}', '', str(x)))
```

```python
df2['cleaned_text'].head()
```

```python
## Tokenizing Articles

def create_tokens_articles(sentence):
    list_tokens=list(gensim.utils.simple_preprocess(str(sentence), deacc=True, min_len=2, max_len=15))  # deacc=True removes punctuations
    list_tokens = [word for word in list_tokens if word not in stop_words]
    list_tokens = [word for word in list_tokens if not word.isnumeric()]
    return list_tokens
```

```python
%%time
df2['cleaned_text_tokens']=df2['cleaned_text'].apply(lambda x: create_tokens_articles(x))
```
