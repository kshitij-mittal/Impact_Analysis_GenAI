```python
%%time

df_news_final_project = pd.read_parquet('https://storage.googleapis.com/msca-bdp-data-open/news_final_project/news_final_project.parquet', engine='pyarrow')
df_news_final_project.shape 
```

```python
# Convert date to a datetime function
df_news_final_project['date']=pd.to_datetime(df_news_final_project['date'])
df_news_final_project.info()
```

```python
df_news_final_project.isna().sum()
# No NAs in the dataset
```

## Cleaning the Text

```python
df_news_final_project_v2=df_news_final_project.copy()
# First cleening the news articles

# Removing mentions and urls
df_news_final_project_v2['cleaned_text']=df_news_final_project_v2['text'].apply(lambda x: re.sub(r'(?:\@|http?\://|https?\://|www)\S+','',str(x)))

# Removing new lines
df_news_final_project_v2['cleaned_text']=df_news_final_project_v2['cleaned_text'].apply(lambda x: re.sub(r'(?:[\n]+)',' ',str(x)))
df_news_final_project_v2['cleaned_text']=df_news_final_project_v2['cleaned_text'].apply(lambda x: re.sub(r'(?:[\t]+)',' ',str(x)))

# Removing hashtags
df_news_final_project_v2['cleaned_text']=df_news_final_project_v2['cleaned_text'].apply(lambda x: re.sub(r'(?:\B#\w*[a-zA-Z]+\w*)','',str(x)))
```

```python
# Removing website names
df_news_final_project_v2['cleaned_text']=df_news_final_project_v2['cleaned_text'].apply(lambda x: re.sub(r'\b\w+\.(com|net|org|edu|gov|io|co)\b','',str(x)))
```
